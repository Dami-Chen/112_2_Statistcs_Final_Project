# Data Preprocessing

```python
import pandas as pd
```

```python
def basketball_reference_data_preprocessing(df):
    pd.options.mode.chained_assignment = None
    df_grouped = df.groupby('Player')
    fractions = []
    for player in df_grouped:
        if player[1].shape[0] > 1:
            extract = df[df['Player'] == player[0]]
            df = df.drop(df[df['Player'] == player[0]].index)
            max_game_played = 0
            main_pos = ''
            main_team = ''
            for i in range(len(extract)):
                if extract['Tm'].iloc[i] == 'TOT':
                    continue
                if max_game_played < extract['G'].iloc[i]:
                    max_game_played = extract['G'].iloc[i]
                    main_pos = extract['Pos'].iloc[i]
                    main_team = extract['Tm'].iloc[i]
            accumulated = extract[extract['Tm'] == 'TOT']
            accumulated['Pos'].iloc[0] = main_pos
            accumulated['Tm'].iloc[0] = main_team
            fractions.append(accumulated)
    fractions.append(df)
    result = pd.concat(fractions).reset_index().drop(columns = ['index'])
    return result
```

## Preprocessing traditional data

```python
traditional = pd.read_csv('./player_stats.csv')
traditional = traditional.drop(columns = ['Rk', 'Player-additional'])
traditional
```

```python
traditional = basketball_reference_data_preprocessing(traditional)
traditional
```

## Preprocessing advanced data

```python
advanced = pd.read_csv('./advanced.csv')
advanced = advanced.drop(columns = ['Rk', 'Player-additional'])
advanced.dropna(how = 'all', axis = 1, inplace = True)
advanced
```

```python
advanced = basketball_reference_data_preprocessing(advanced)
advanced
```

```python
advanced = advanced.drop(columns = ['Pos', 'Age', 'Tm', 'G', 'MP'])
advanced
```

## Preprocessing salary data

```python
salary = pd.read_excel('./player_salary.xlsx')
salary = salary[['PLAYER', '2023/24']]
salary = salary.rename(columns = {
    'PLAYER': 'Player',
    '2023/24': '2023-24_Salary'
})
salary
```

```python
print(salary['Player'].value_counts().value_counts())
print(len(salary['2023-24_Salary'].dropna()))
```

## Merging the three datasets

```python
performance = traditional.merge(advanced, how = 'inner', on = 'Player')
performance
```

```python
performance_with_salary = performance.merge(salary, how = 'inner', on = 'Player')
performance_with_salary
```

```python
performance_with_salary = performance_with_salary.sort_values('2023-24_Salary', ascending = False).reset_index().drop(columns = ['index'])
performance_with_salary
```

```python
performance_with_salary.to_csv('./performance_with_salary.csv', index = False)
```
